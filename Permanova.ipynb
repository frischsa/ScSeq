```python
import numpy as np
import pandas as pd
from skbio.stats.distance import permanova, DistanceMatrix
from scipy.spatial.distance import pdist, squareform
```

```python
df = pd.read_excel("/Users/mimi/Desktop/Fantasiezahlen.xlsx", sheet_name="Sheet1")
```

```python
df
```

```python
data = df.iloc[:, 1:]
groups = df["Methode"].tolist()
```

```python
# Berechne die Distanzmatrix (z.B. mit Bray-Curtis-Distanz)
dist_array = pdist(data.values, metric='braycurtis')
dist_matrix = DistanceMatrix(squareform(dist_array), ids=[str(i) for i in range(len(data))])
```

```python
# Führe PERMANOVA mit 999 Permutationen durch
result = permanova(distance_matrix=dist_matrix, grouping=groups, permutations=999)
print(result)
```
